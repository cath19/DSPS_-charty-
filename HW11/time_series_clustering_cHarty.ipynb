# cluster the data in 2 to 10 clusters and print the total variance for each cluster. 
# Plot the total variance as a function of number of clusters

```
from sklearn import cluster
```

```
nmaxc = 10
nc = []
for i in range(1, nmaxc):
  c = cluster.k_means(X, i)
  nc.append(c[2])
  print("i.c. variance", c[2])
pl.plot(range(1, nmaxc), nc)
pl.xlabel("number of clusters")
pl.ylabel("total intracluster variance");
pl.title('Total Intracluster Variance vs Number of Clusters')
```

**Figure 3**: The total intracluster variance vs the number of clusters. Note where the elbow appears.

pick some number of clusters and

- print the number of time series in each cluster
- for each cluster plot the cluster center and each time series that belong to that center in the same plot
- print the names of the countries in the smallest cluster

```
# making sure I am clustering along the right axis: 
# each cluster center shoul de as long as the timeline: 59
X.shape, c[0].shape
```

```
clusters = cluster.KMeans(n_clusters=4).fit(X)
counts = []
print('The number of time series in each cluster:')
for i in range(np.unique(clusters.labels_).size):
  count = (clusters.labels_ == i).sum()
  counts.append([i + 1, count])
  print('Cluster %d:\t%d features' % (i + 1, count))

cmap = {0:'C0', 1:'C1', 2:'C2', 3:'C3'}
[pl.plot(X[i], c = cmap[clusters.labels_[i]]) for i in range(X.shape[0])];
pl.xlabel("year")
pl.xticks(ticks=range(0, 70, 10), 
          labels=["%d"%i for i in range(1960, 2030, 10)])
pl.ylabel("normalized + standardized population");
pl.title('Normailized and Standardiazed Population Data from 1960 - 2019')
```

**Figure 4**: Same plot as Figure 2, but it has now been clustered into 4 clusters.

```
for i in range(len(clusters.cluster_centers_)):
  for j in range(len(clusters.labels_)):
    pl.title('Cluster %d' % (i + 1))
    if clusters.labels_[j] == i:
      pl.plot(X[j], c = cmap[i]);
      pl.plot(clusters.cluster_centers_[i], c = 'k', linewidth = 1.5);
      pl.xticks(ticks=range(0, 70, 10), labels=["%d"%i for i in range(1960, 2030, 10)]);
      pl.xlabel('year')
      pl.ylabel("normalized + standardized population");
  pl.figure();
```

**Figure 5**: A close up of each of the 4 clusters.
